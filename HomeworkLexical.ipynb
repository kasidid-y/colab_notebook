Clean Text

```python
x = re.sub("ำ","า", text)
print(x)
text = x
```

Text to sentences

```python
sen_token = sent_tokenize(text, engine="whitespace+newline")
print(sen_token)
sen_token01 = sent_tokenize(text, engine="crfcut")
print(sen_token01)
```

#Senteces to words with newmm and newmm-safe tokenizer

```python
# import pythainlp
from pythainlp.tokenize import Tokenizer
from pythainlp.corpus.common import thai_words
from pythainlp.util import dict_trie
```

```python
# print(thai_words())
```

```python
print("newmm")
for word in sen_token:
  tokens = word_tokenize(word, engine="newmm")
  print(tokens)
print("-----\n")
print("newmm-safe")
for word in sen_token:
  tokens = word_tokenize(word, engine="newmm-safe")
  print(tokens)
print("-----\n")
```

# Adding a custom words and Tokenize with custom dictionaries


```python
custom_words_list = set(thai_words())
custom_words_list.add('เสริมศักดิ์')
custom_words_list.add('พงษ์พานิช')
custom_words_list.add('ลาบูบู้')
custom_words_list.add('มาสคอต')
custom_words_list.add('สุวรรณภูมิ')
trie = dict_trie(dict_source = custom_words_list)

for word in sen_token:
  tokens = word_tokenize(word, engine="newmm", custom_dict=trie)
  print(tokens)
```
